# Fluid mechanics of evaporating <br> sessile drops

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# LaTeX settings
plt.rc('text', usetex=True)
plt.rc('font', family='serif')  
```

## Defining the problem

#### Inputs
What can we measure?
<ul>
<li>radius (R)</li>
<li>contact angle (a)</li>
<li>time (t)</li>
</ul>

#### Outputs
What do we want to output?
<ul>
<li>height (h) at specific radius (r), at a time (t)</li>

### An equation to describe height at a unisymmetrical position (r) as a function of time, across a drop of radius (R)

For an individual drop at a time (t) with radius (R) and contact angle (a):
<br>&nbsp;</br>
$$ h(r,t) = \sqrt \frac{R^2(t)}{sin^2 \theta (t)} - r^2-R(t)\cot\theta(t)$$

```python
def h(r, t, R, a):
    a_rad = math.radians(a)
    h = math.sqrt(((R**2)/((math.sin(a_rad))**2))- r**2) - R * (1 / math.tan(a_rad))
    return h

h(10, 10, 20, 1)
```

### Height across the x-radius of the drop

```python
height = []
r = range(-10,10+1)

for i in r:
    hx = h(i, 60, 10, 40)
    height.append(hx)

plt.plot(r,height)
plt.xlabel('Radius')
plt.ylabel('Height')
plt.title('Height across droplet')
plt.show()
```

### Plotting heights for multiple contact angles

```python
r = range(-10,10+1)
ca = range(10, 100, 10)

for j in ca:
    height = []
    for i in r:
        hx = h(i, 60, 10, j)
        height.append(hx)
    plt.plot(r, height, label=j)
    
# plt.plot(r,height)
plt.xlabel('Radius')
plt.ylabel('Height')
plt.ylim(0,10)
plt.title('Height across droplet')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

### Height with changing contact angle

```python
height = []
r = range(1, 180)

for i in r:
    hx = h(0, 60, 10, i)
    height.append(hx)

plt.plot(r,height)
plt.xlabel('Radius, degrees')
plt.ylabel('Height')
plt.title('Height across droplet')
plt.show()
```

### Contact angle as a function of radius and h0
This is a linear way of representing the peaks of the bands seen in the graph of heights given different contact angles

```python
def contact_angle(R, h0):
    CA = math.degrees(2 * math.atan(h0 / R))
    return CA

x = range(1,10+1)
y = []
for i in x:
    yi = contact_angle(10, i)
    y.append(yi)
    
plt.title(r'Changing contact angle with maximum \newline drop height at a constant drop radius of 10')
plt.xlabel(r'Max height, h0')
plt.ylabel(r'Contact angle, degrees')

plt.plot(x,y)
plt.show()
```

### h0 as a function of contact angle and radius
This is a linear way of representing the peaks of the bands seen in the graph of heights given different contact angles

```python
# R.tan(0/2) = h0

def max_height(R, a):
    h0 = R * math.tan(math.radians(a)/2)
    return h0

#max_height(10, 60)

R = 10
c_angle = range(0, 92, 2)
m_height = []

for i in c_angle:
    hh = max_height(R, i)
    m_height.append(hh)

#print(m_height)
plt.plot(c_angle, m_height)
plt.xlabel('Contact angle, degrees')
plt.ylabel('Max. height')
plt.title('Change in max height with increasing contact angle')
plt.show()
```
